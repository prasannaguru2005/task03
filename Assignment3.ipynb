1.Employee Performance Bonus Eligibility

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

top_score = max(employees.values())

for name in employees:
    if employees[name] == top_score:
        print(name, end=", ")

print("(Score:", top_score, ")")
```

2.Search Query Keyword Analysis

```python
text = "Buy mobile phone buy phone online"
text = text.lower()
for ch in ".,!?":
    text = text.replace(ch, "")
words = text.split()
count = {}
for word in words:
    count[word] = count.get(word, 0) + 1
result = {}
for word in count:
    if count[word] > 1:
        result[word] = count[word]
print(result)
```

3.Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = []
for i in range(len(sensor_readings)):
    if sensor_readings[i] % 2 == 0:   
        valid_readings.append((i, sensor_readings[i]))
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

 4.Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
domain_count = {}
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1
total = len(emails)
for domain in domain_count:
    percent = (domain_count[domain] / total) * 100
    print(f"{domain}: {int(percent)}%")
```

5.Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
total_sum = sum(sales)
n = len(sales)

for i in range(n):
    avg = (total_sum - sales[i]) / (n - 1)
    threshold = avg * 1.3
    if sales[i] > threshold:
        print(f"Day {i+1}: {sales[i]}")
```
